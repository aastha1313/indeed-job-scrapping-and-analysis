```python
import pandas as pd

df = pd.read_excel("indeed_jobs.xlsx")

most_common_title = df['Title'].value_counts().idxmax()
count = df['Title'].value_counts().max()

print(f"most common job title is: '{most_common_title}' with {count} listings")
```

```python

df = pd.read_excel("indeed_jobs.xlsx")


location_counts = df['Location'].value_counts().reset_index()
location_counts.columns = ['Location', 'Number of Listings']

location_counts.head()
```

```python

title_counts = df['Title'].value_counts().reset_index()
title_counts.columns = ['Job Title', 'Number of Listings']

title_counts.head()
```

```python
import matplotlib.pyplot as plt
```

```python

location_counts = df['Location'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Job Listings by Location")
plt.axis('equal')  
plt.show()
```

```python

title_counts = df['Title'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(title_counts, labels=title_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Job Listings by Job Title")
plt.axis('equal')
plt.show()
```
